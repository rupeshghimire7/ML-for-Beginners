# Classification

Classification means to categorize. In machine learning, we're categorizing given sets of data into classes. Classes are also known as target, categories or labels. Classification can done in both supervised and unsupervised models.

**for e.g** Classifiying if the message received is spam or not

# Types of Classification

### 1. Binary Classification

  Our model needs to predict **one** of the given **two** classes for given set of features or input variables. Hence, We have only 2 labels for our classifiers to predict.

Popular algorithms that  can be used are:
  * Logistic Regression
  * Na√Øve Bayes
  * K-Nearest Neighbours
  * Decision Tree
  * Random Forest
  * Support Vector Machine

**for e.g.**
  * Email Spam Detection (Spam or Not)
  * User Shopping Behavior (Buy or not)
  * Breast Cancer Detection (Yes or No)

### 2. Multi Class Classification
  Our model needs to predict **one** of the given **many** classes for given set of features or input variables. Hence, We have more than 2 labels for our classifiers to predict. Binary Classification can be regarded as special case of Multi Class Classification.

Popular algorithms that  can be used are:
  * Artificial Neural Networks (ANN)
  * One Vs All
  * Random Forest
  * Gradient Boosting
  * Support Vector Machine
  * Logistics Regression
  * Xgboost
  
**for e.g.**  
  * Digit Recognization ( 0-9)
  * Fingerprint Classification (Each individual has separate fingerprint so 1 from 7 billion)

### 3. Multi Label Classification
  In multilabel classification , our classifier needs to classify **one or more than one labels** from given set of labels for given set of input features. Hence, Each row may consist of multi labels for us to predict.

Popular algorithms that  can be used are:
  * Artificial Neural Networks (ANN) Approach
  * One Vs Rest Classifier   
  * Problem Transformation -
    * Binary Relevance
    * Classifier Chains
    * Label Powersets
  * Adapted Learning
    * Multi KNN (MLkNN)
    * Multi SVM (MLTSVM)
  * Ensemble Methods
    * Multi Decision Tree
    * Multi Gradient Boosting
    * Multi Random Forest

(Ref: [Scikit.multilearn](http://scikit.ml/api/skmultilearn.html))

**for e.g.**  
  * Object Detection in  Images
  * Hashtag Detection in Social Media Sites

### **STEP-3** (Modeling)
Modelling consists of three important things we need to do with our model.
1. **Instantization**

    The first process is loading our model. Our model can be custom made code or can be load from one of the machine learning library such as Scikit-Learn. Then after, we create an instance of the model class we loaded.

2. **Fit**
  
  We pass our features set(x_train) and target set(y_train)  to instance of the model so that our  models learns from it. Simply to say, Fitting means learning.Fit function adjusts weights according to data values so that the model can achieve better accuracy.

3. **Predict**

  For the given set of input data, predict is done to map it into the target data. We tend to pass the test set data into predict method in order to compare the original target values(y_test) and predicted target value by the model(y_pred).

Lets, go with an example of Spam vs Not Spam which is an binary classification example.
  
                                                                        
| N=100      | Predicted True | Predicted False     |
| :---        |    :--:   |          ---: |
| **Actual True**      | TP = 40       | FN = 5    |
| **Actual False**   | FP = 10        |   TN = 45   |

Above is some hypothetical situation of a classifier predicting 100 message from a datasets whether the message is spam or not.

a . **True Positive(TP)**  

When your predictor predicts true(spam) and it the actual data is also true(spam), then it is called True Positive.

b. **True Negative(TN)**

When your predictor predicts false( not spam) and it the actual data is also false(not spam), then it is called True Negative.

c. **False Positive(FP)**

When your predictor predicts true(spam) but the actual data is false(not spam) , then it is called False Positive. It is also called *Type-I Error*.

d. **False Negative(FN)**

When your predictor predicts false(not spam) but the actual data is true(spam) , then it is called False Negative. It is also called *Type-II Error*.

**FN** is more dangerous kind of error in medical world. for e.g. You are predicting whether a patient has breast cancer or not. And, Your models predicts false (Not Breast Cancer) when the patient actually has breast cancer (True). As you can see, the prediction is hazardous in respect to patient.

But, In case of **FP**, If your model predicts you have breast cancer (True) when you  actually don't have breast cancer(False) then it is not so dangerous comparing to the case of False Negative.

But, FP and FN are morely interchangeably with respect to the model. For say , a programmer may program having breast cancer as negative and not having as positive. Hence, In that case above  situation is reversed.





 2. **Accuracy**

 Confusion matrix gives us other form of measuring metrics too.
 Accuracy answers us `Overally,How often is the classifier correct?`

 $ Accuracy = \frac{TP + TN}{Total}   $
$ = \frac{40+45}{100}$ $ = \frac{85}{100}$ $ = 0.85$

3. **Precision**

 Precision answers `When it predicts yes, how often is it correct?`

 OR

 `Out of all the positive classes we have predicted correctly, how many are actually positive`

  $ Precision =\frac{TP }{Predicted Yes} $ $ = \frac{TP }{TP + FP}   $
$ = \frac{40}{40+10}$ $ = \frac{40}{50}$ $ = 0.8$

4. **Recall (Sensitivity)**

 Recall answers `When it's actually yes, how often does it predict yes?`

 OR

 `Out of all the positive classes, how much we predicted correctly.`

  $ Recall =\frac{TP }{Actual Yes} $ $ = \frac{TP }{TP + FN}   $
$ = \frac{40}{40+5}$ $ = \frac{40}{45}$ $ = 0.89$

5. **F1 Score**

 Sometimes, It is difficult to compare two models with low precision and high recall or vice-versa.  F1 score is the harmonic mean or weighted average of Recall and Precision.

 $ f1-score =2 * \frac{ Precision * Recall}{Precision + Recall} $
$ = 2 * \frac{0.8 * 0.89 }{0.8 + 0.89}$ $ = 2* \frac{0.712}{1.69}$ $ = 0.84$


#Balanced vs Imbalanced Datasets

In **Balanced Datasets**, there is considerably equal amount of Positive and Negative target values. (i.e 50%). Or, in case of multi class classification, there are equal amount of target values present in the datasets.

If there is the very high different between the positive values and negative values. Then we can say our dataset in **Imbalance Dataset**.
